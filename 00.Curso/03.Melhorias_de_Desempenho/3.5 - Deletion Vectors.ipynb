## Conceito

Os vetores de exclusão (deletion vectors) são um recurso de otimização de armazenamento que você pode habilitar em tabelas Delta Lake. Por padrão, quando uma única linha em um arquivo de dados é atualizada ou excluída, todo o arquivo Parquet que contém esse registro precisa ser reescrito. Com os vetores de exclusão habilitados para a tabela, as operações de DELETE, UPDATE e MERGE criam pequenos arquivos chamados "vetores de exclusão" para marcar as linhas existentes como removidas ou alteradas, sem a necessidade de reescrever o arquivo Parquet original. As leituras subsequentes na tabela resolvem o estado atual aplicando as exclusões indicadas pelos vetores à versão mais recente da tabela.

Portanto, os vetores de exclusão indicam as alterações nas linhas como "soft-deletes" (exclusões lógicas) que modificam logicamente os arquivos de dados Parquet existentes na tabela Delta Lake. Essas alterações são aplicadas fisicamente quando um dos seguintes eventos faz com que os arquivos de dados sejam reescritos:

    Um comando OPTIMIZE é executado na tabela.

    A compactação automática (auto-compaction) dispara a reescrita de um arquivo de dados que possui um vetor de exclusão.

Como habilitar os vetores de exclusão

Os vetores de exclusão já vêm habilitados por padrão quando você cria uma nova tabela usando um SQL Warehouse ou o Databricks Runtime 14.1 ou superior. Para habilitar ou desabilitar manualmente o suporte aos vetores de exclusão em qualquer tabela ou view Delta, utilize a propriedade de tabela 

`delta.enableDeletionVectors`
