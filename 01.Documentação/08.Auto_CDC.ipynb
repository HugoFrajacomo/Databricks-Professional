## Syntax SQL

```python
%sql
CREATE FLOW flowname AS
AUTO CDC INTO my_table
FROM STREAM(cdc_my_table)
KEYS(key_fild)
APPLY AS DELETE WHEN operation_type = "DELETE"
SEQUENCE BY sequence_field
COLUMNS * EXCEPT (operation_type, sequence_field)
STORED AS SCD TYPE 1;
```

### Express達o antiga (n達o recomendada)

```python
%sql
APPLY CHANGES INTO target_table
FROM STREAM(cdc_my_table)
KEYS(key_fild)
APPLY AS DELETE WHEN operation_type = "DELETE"
SEQUENCE BY sequence_field
COLUMNS * EXCEPT (operation_type, sequence_field)
STORED AS SCD TYPE 1;
```
