**Singleplex:** One-to-One Mapping: Fonte de dados individuais (tabela em csv, sql, json etc) que são mapeadas uma a uma para suas próprias tabelas. 

![](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRyD1NeMbSidugfYqvmnxL3bS7vM_eKE_Kszg&s)

**Multiplex:** Many-to-One Mapping: Adiciona multiplas fontes em uma unica tabela bronze unificada. Em vez de criar uma tabela para cada origem, os dados chegam a um repositório cental que possui uma coluna identificadora (Nome do Topico ou Origem do Arquivo) e uma coluna de carga útil(payload normalmente em JSON)
