## Comandos por SQL

```python
%sql
-- Criando catalogo
CREATE CATALOG IF NOT EXISTS Projeto;

--Criação de Schemas
CREATE SCHEMA IF NOT EXISTS Bronze;
CREATE SCHEMA IF NOT EXISTS Silver;
CREATE SCHEMA IF NOT EXISTS Gold;

-- Criação de volumes
CREATE VOLUME IF NOT EXISTS Staging_data;

-- Criando uma tabela gerenciada na camada Bronze
CREATE TABLE IF NOT EXISTS Projeto.Bronze.Logs_Brutos (
  id_evento STRING,
  data_evento TIMESTAMP,
  payload STRING
) USING DELTA;

-- Criando uma tabela externa na camada Gold (necessita de External Location configurado)
CREATE TABLE IF NOT EXISTS Projeto.Gold.Faturamento_Mensal
LOCATION 'abfss://container@account.dfs.core.windows.net/gold/faturamento'
AS SELECT * FROM Projeto.Silver.Vendas_Tratadas;

-- Criando uma função
CREATE OR REPLACE FUNCTION soma (a INT, b INT)
RETURNS INT
COMMENT 'Função que retorna a soma de dois inteiros.'
RETURN a + b;

-- Concedendo acesso de leitura para um grupo de analistas na camada Gold
GRANT USAGE ON CATALOG Projeto TO `grupo_analistas`;
GRANT USAGE ON SCHEMA Projeto.Gold TO `grupo_analistas`;
GRANT SELECT ON SCHEMA Projeto.Gold TO `grupo_analistas`;

-- Concedendo permissão total para um engenheiro de dados em um schema específico
GRANT ALL PRIVILEGES ON SCHEMA Projeto.Silver TO `usuario_engenheiro@empresa.com`;
```
