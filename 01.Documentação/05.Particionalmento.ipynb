## Particionamento 
Uma partição é um subconjunto de linhas que compartilham o mesmo valor para colunas de particionamento predefinidas.

Ex: Uma tabela contendo id, nome, mês, ano. Poderia ser participana por ano e agrupar arquivos pelo ano.



```python
%sql
--Exemplo de código
CREATE TABLE my_table(id INT, name STRING,  year INT, month INT)
PARTITIONED BY (year)
```

Desta forma quando executarmos uma consulta delimitando o ano primeiro ele vai procurar o bloco correspondente ao ano e ignorar os dados dos blocos dos outros anos

### Particionamento por mais de uma coluna

```python
%sql
CREATE TABLE my_table(id INT, name STRING,  year INT, month INT)
PARTITIONED BY (year, month)
```

## Escolhendo a coluna a ser particionada

- Total de valores presentes em uma coluna:

Campos de baixa cardinalidade devem ser usados para particionamento.

- Total de registros que compartilham um determinado valor em uma coluna:

As partições devem ter pelo menos 1 GB de tamanho.

- Registros com um determinado valor continuarão chegando indefinidamente.

Os campos de data e hora podem ser usados para particionamento.


IMPORTANTE: tabelas pequnas particionadas tem performance inferior a não particionadas
