## Databricks-Databricks Sharing protocol

Permite compartilhamento do catalog entre clientes do databricks, como por exemplo:

- Coleções de tabelas, volumes, views e notebooks
- Utiliza built-in Authentication

```python
%sql
-- Permite compatirlhar history sharing
ALTER SHARE bookstore ADD TABLE orders WITH HISTORY;
```

## Databricks open sharing protocol

- Compartilha dados com usuários em qualquer plataforma computacional
- Apenas compartilha tabelas
- Requer autenticação via tokens ou por OpenID Connect (OIDC) Federation

## Custos

Delta Sharing não requere duplicação de dados, desta forma não há maiores custos referente a armazenamento

O que aumenta o custo: 

- Serviços cross-plataformas ou cross-regiões é cobrado uma taxa.

Como evitar o custo:

- Clonar o dado para a região antes de compartilhar
- Ou utilizar outra forma de compartilhar como por exemplo cloudflare R2 que não aplica taxas

## Limitações

Todos os dados fornecidos com data sharing são apenas read only. As tabelas também tem que estar em formato Delta Table
