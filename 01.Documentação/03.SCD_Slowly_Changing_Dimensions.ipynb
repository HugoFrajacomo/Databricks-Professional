Slowly Changing Dimensions (SCD) são técnicas de modelagem em Data Warehousing utilizadas para gerenciar e armazenar mudanças nos dados das dimensões ao longo do tempo, em vez de simplesmente sobrescrever as informações antigas. Em arquiteturas robustas de Lakehouse com Databricks, essas técnicas são fundamentais para manter a integridade histórica, permitindo que analistas e cientistas de dados realizem consultas retroativas. Sem o uso de SCD, ao alterar o endereço de um cliente, por exemplo, todos os registros de vendas passadas pareceriam ter ocorrido no novo local, o que comprometeria a precisão de relatórios históricos e auditorias de governança.

Existem diversos tipos de SCD, sendo os mais comuns o Tipo 1, onde os dados são sobrescritos e não há histórico, e o Tipo 2, que cria uma nova linha para cada alteração, utilizando colunas de controle como datas de validade (início e fim) e flags de registro atual.
