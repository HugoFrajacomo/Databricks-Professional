## Como definir a função em SparkSQL

```python
window(time_column, windows_size, slide_interval, starting_offiset)

#onde: 
#   time_column: Coluna que vai ser usada para calcular o tempo
#   windows_size: Define a duração de cada janela.
#   slide_interval: Define a frequência com que uma nova janela é iniciada.
#   starting_offiset: Desloca todas as janelas para frente ou para trás no tempo.
```

### Exemplo

- window_size = "10 minutes" -> Cada janela dura 10 minutos
- slide_interval = "10 minutes" -> A cada 10 minutos executa uma nova janela
- Starting_offiset = "3 minutes" -> Iniciam 3 minutos depois do alinhamento natural
