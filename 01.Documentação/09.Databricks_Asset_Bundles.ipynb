## Asset Bundles (DABs)

Permite a implementação de workflows através de multiplos ambientes

- Dev Workspaces
- Staging Workspaces
- Prod Workspaces

Pode ser integrado ao git, conforme faz um commit & push ele aciona a triguer que valida, faz deploy e roda a execução.  tanto em branch de desenvolvimento quando em produção

## Comandos CLI

```python
databricks bundle init [template-name] #Criação de um novo bundle
databricks bundle validate #Valida se os arquivos estão corretos
databricks bundle deploy -t <target-name> #Deploy do bundle para o target
databricks bundle run <job-name> -t <target-name> #Execução de um job dentro do bundle
```
